```python
import pandas as pd
url = "https://minio.lab.sspcloud.fr/mligeret1/ValeursFoncieres-2024.txt"
df = pd.read_csv(url, sep="|", low_memory=False)
```

```python
df.head()
```

```python
print(df.columns)
colonne_pertinentes = ["Valeur fonciere", "Date mutation",'No voie', 'B/T/Q', 'Type de voie', 'Code voie', 'Voie', 'Code postal',
       'Commune','Surface reelle bati', 'Nombre pieces principales','Type local', "Identifiant local", "Prix au m2"]
```

```python
df["Valeur fonciere"] = df["Valeur fonciere"].str.replace(",",".")
```

```python
df = df.astype({"Valeur fonciere":"float64"})
```

```python
print(df.dtypes)
```

```python
df["Prix au m2"] = df["Valeur fonciere"]//df["Surface reelle bati"]
df.loc[:,colonne_pertinentes].tail(60)
df_clean = df.dropna(subset=["Prix au m2"])


m2_par_rue = df_clean.groupby(["Commune","Voie"])["Prix au m2"].mean().to_frame()
m2_par_rue = m2_par_rue["Prix au m2"].sort_values(ascending=True).to_frame()
m2_par_rue.head()
```

```python
m2_par_rue.loc[(m2_par_rue["Surface au m2"]< 1000000) & (m2_par_rue["Surface au m2"]> 19000)]
```
