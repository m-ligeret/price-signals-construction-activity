```python
# Viewing columns and selecting relevant ones
print(df.columns)
colonnes_pertinentes = ["No disposition","Valeur fonciere", "Date mutation",'No voie', 'B/T/Q', 'Type de voie', 'Code voie', 'Voie', 'Code postal',
       'Commune','Code commune','Surface reelle bati','Surface terrain', 'Nombre pieces principales','Type local', "Identifiant local"]
```

```python
# Defining the new dataframe that we'll use
df_pertinent = df[colonnes_pertinentes].copy()
```

```python
# Checking importance of null value for transaction value
df_pertinent["Valeur fonciere"].isna().sum()
```

```python
# Removing those values as well as duplicate lines
filter_na = df_pertinent["Valeur fonciere"].isna()
df_removed = df_pertinent[~filter_na].drop_duplicates() # New clean dataframe
```

```python
# Cleaning of price and construction of price by m2
df_removed["Valeur fonciere"] = df_removed["Valeur fonciere"].str.replace(",",".",regex=False) # Replacing , by . 
df_removed["Valeur fonciere"] = df_removed["Valeur fonciere"].astype(float).round().astype(int) # Converting to int (removing decimals if needed)
df_removed["Prix au m2"]=df_removed["Valeur fonciere"]/df_removed["Surface terrain"]
```

```python
#Checking that we have dates for each transaction
df_removed["Date mutation"].isna().sum()
```

```python
# Grouping lines by transaction value and date, to aggregate lines corresponding to the same transaction
df_groupby = df_removed.groupby(['Valeur fonciere', 'Date mutation'])['Voie'].transform('nunique')
filter = (df_groupby>1) & (df_removed["Date mutation"]=='09/01/2024')
df_removed[filter]
```
